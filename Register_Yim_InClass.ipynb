# In-class Gradient Descent Stuff

```python
#loading all my libraries lol, definitely don't need all these
import IPython
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib
import sklearn
import matplotlib.pyplot as plt  
```

```python
x = np.array([1,2,3])
y= np.array([1,2,3])

with plt.xkcd():
    plt.scatter(x, y, alpha=0.5,color="red")
    plt.title("THE DUMBEST PLOT IN THE WORLD")
    plt.show()
#hypothesis being that y=x

#a linear model would be y= alpha + Beta*x

```

```python
#how to penalize for making mistakes?
# J(alpha, Beta)
# minimize that cost function!

def make_linear_hypothesis(alpha,beta,length):
    hyp = []
    for i in range(0,length):
        hyp.append(beta*i+alpha)
    return hyp

hyps = []
betas = []

#generate out hypotheses (100 of them for now)
for i in range(100):
    bet = np.random.random()*np.random.randint(-10,10);
    betas.append(bet)
    h = make_linear_hypothesis(0,bet,3)
    hyps.append(h)
#print(hyps)
    

with plt.xkcd():
    plt.scatter(x, y, alpha=0.5,color="red")
    plt.title("THE DUMBEST PLOT IN THE WORLD")
    for a in hyps:
        plt.plot(x, a)
    
    plt.show()

    
def eval_cost(true,predicted):
    cost = np.sqrt(np.sum(((np.array(predicted)-np.array(true))**2)/len(predicted)))
    return cost

#so I've generated some hypotheses, h1,h2,h3,and h4

costs = []
for h in hyps:
    c= eval_cost(y,h)
    costs.append(c)
print(costs)
    

        
```

```python
with plt.xkcd():
    plt.scatter(betas,costs)
    plt.annotate(
        'MINIMA',
        xy=(1.5, 1), arrowprops=dict(arrowstyle='->'), xytext=(1, 5))

    plt.title("LOOK AT THAT CONVEX FUNCTION AW YEAH")
    plt.show()
```

# Here's the funky hypothesis fitting aka what if we had the worst hypothesis ever

```python
def make_funky_hypothesis(param1, param2, param3,length):
    hyp = []
    for i in range(1,length):
        hyp.append(((param1+param2/i)**param3)/param1) #like, what if for some reason this is actually the generating function????
    return(hyp)
h = make_funky_hypothesis(9,9,2,10)

hyps = []
params = []

#generate out hypotheses (100 of them for now)
for i in range(100):
    p1= np.random.random()*np.random.randint(1,10);
    p2 = np.random.random()*np.random.randint(-10,10);
    p3= np.random.random()*np.random.randint(-10,10);
    params.append([p1,p2,p3])
    h = make_funky_hypothesis(p1,p2,p3,10)
    hyps.append(h)
#print(hyps)

l = np.linspace(1,10,9)

with plt.xkcd():
    
    plt.title("THE SECOND DUMBEST PLOT IN THE WORLD")
    for a in hyps:
        plt.ylim(-5,100)
        plt.plot(l, a)
    
   
    plt.show()
    
    plt.scatter(l,h,color="black")
    plt.title("THE 'REAL' DATA")
    plt.show()
```
